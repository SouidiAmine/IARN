# Entrainement 
dans cette partie nous allons calculer les parametres de la disribution gaussienne de chaque caracteristique.

```python
plt.scatter(X[:,0],X[:,1], marker ='x')
#plt.scatter(X1[:,0],X1[:,1], marker ='o')
```

Mettons un lien avec ce que nous avons appris en regression linéaire

```python
m = np.mean(X, axis=0)
v = np.var(X, axis=0)
```

```python
m
```

```python
def calculProba(X,m,v):
    return (1 / (np.sqrt(2 * v * np.pi))) * np.exp(- np.power(X - m, 2) / (2 * v))
```

```python
f1_scores = []
epsilons = []
proba = calculProba(X, m, v)
yproba = proba[:,0]*proba[:,1]
for epsilon in np.arange(0, 1, 0.0001):
    y_pred = np.ones(yproba.shape)
    y_pred[yproba >= epsilon] = 0
    #f1_scores.append((epsilon, f1_score(y, y_pred)))
    f1_scores.append(f1_score(y, y_pred))
    epsilons.append(epsilon)
```

```python
print('Score = ',f1_scores[np.argmax(f1_scores)])
epsilon = epsilons[np.argmax(f1_scores)]
#epsilon = seuil qui on vas tester et clasifier avac lui
print ('epsilon = ',epsilon)
```

# Multivariate Guassien Distribution 


```python
a = np.array([[ 0.33017929,  1.10756915,  1.25937971],
       [-1.30714955,  0.21901611, -0.22115456]])
B = np.linalg.pinv(a)
print(a)
B
```

```python
np.array([[-0.00296532, -0.72474294],
       [ 0.41571182,  0.2289517 ],
       [ 0.42921884, -0.01134267]])
```

```python
import numpy as np


def est_gaus_par(X):
    mu = np.mean(X,axis=0)
    sig =[np.std(X,axis=0)[0],np.std(X,axis=0)[1]]
    return mu,sig

mu,sigma = est_gaus_par(X)
print("sigma = ",sigma)
def est_mult_gaus(X,mu,sigma):
    m = len(mu)
    sigma2 = np.diag(sigma)
    X = X-mu.T
    p = 1/((2*np.pi)**(m/2)*np.linalg.det(sigma2)**(0.5))*np.exp(-0.5*np.sum(X.dot(np.linalg.pinv(sigma2))*X,axis=1))

    return p

p = est_mult_gaus(X, mu, sigma)
p
```

# Verification avec l'implimentation de librery scipy.stats

```python
cov =[[np.std(X,axis=0)[0],0],[0,np.std(X,axis=0)[1]]]
cov
```

```python
from scipy.stats import multivariate_normal

var = multivariate_normal(mean=m, cov=cov)
P_scipy = var.pdf(X)
P_scipy
```
